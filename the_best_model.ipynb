```python
import pandas as pd
```

```python
data = pd.read_csv('/content/churn-bigml-20.csv')
```

```python
data.isnull().sum()
```

```python
from sklearn.ensemble import VotingClassifier
```

```python
data.duplicated().sum()
```

```python
non_numerical_colums=data.select_dtypes(exclude=["number"]).columns.tolist()
non_numerical_colums
```

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()   #traditional machine learning
for i in non_numerical_colums:
  data[i]=LE.fit_transform(data[i])
```

```python
y = data["Churn"]
x = data.drop("Churn",axis=1 )
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()  # Create an instance of StandardScaler
X=scaler.fit_transform(x) # Apply the scaler to your data (x, not X)
#X=scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=10)
pca.fit(x_train)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
# logReg = LogisticRegression(max_iter=200)
# svc = SVC(probability = True)
# rf = RandomForestClassifier()
# dt = DecisionTreeClassifier()
```

```python
logReg = LogisticRegression(C=0.1, max_iter=1000)

DT = DecisionTreeClassifier(max_depth=4, max_features=30, min_samples_split=5)

RF = RandomForestClassifier(criterion='entropy', max_depth=9, max_features=9, min_samples_split=3, n_estimators=30)

from sklearn.naive_bayes import BernoulliNB # Import BernoulliNB
NB = BernoulliNB(alpha=0.05)

svc = SVC(C=1, degree=15, kernel='sigmoid', probability = True )

VC = VotingClassifier(estimators=[('LogisticRegression', logReg), ('DecisionTreeClassifier', DT),
          ('RandomForestClassifier', RF),('BernoulliNB' , NB), ('SVC' , svc)])
```

```python
Models = {'LogisticRegression':logReg,
          'DecisionTreeClassifier':DT,
          'RandomForestClassifier': RF,
          'BernoulliNB' : NB,
          'SVC' : svc,
          'VotingClassifier' : VC
         }
```

```python
# from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, zero_one_loss, recall_score # Import necessary functions

# best_model = None
# best_score = 0.0
# for model in Models.keys():
#     #Model Fitting
#     Models[model].fit(x_train, y_train)
#     print(str(model).center(30))

#     #___________
#     #Prediction
#     y_pred = Models[model].predict(x_test)

#     #___________
#     #metrices
#     CM = confusion_matrix(y_test, y_pred)
#     F1 = f1_score(y_test, y_pred, average='micro')
#     accuracy = accuracy_score(y_test, y_pred, normalize=False)
#     loss = zero_one_loss(y_test, y_pred, normalize=False)
#     recall = recall_score(y_test, y_pred, average='micro')
#     #___________
#     #Score
#     print('Training Score:', Models[model].score(x_train, y_train))
#     print('Testing Score:', Models[model].score(x_test, y_test))

#     #___________
#     if best_score < F1:
#         best_score = F1
#         best_model = model

#     #___________
#     print()
#     print('confusion_matrix\n',CM)
#     print('F1_score', F1)
#     print('Number of right predictions',accuracy)
#     print('Number of wrong predictions',loss)
#     print('Recall Score:', recall)
#     print('_________________________________________________')
```

```python
from sklearn.metrics import confusion_matrix, f1_score, accuracy_score, zero_one_loss, recall_score, precision_score, classification_report

best_model = None
best_score = 0.0

for model in Models.keys():
    # Model Fitting
    Models[model].fit(x_train, y_train)
    print(str(model).center(30))

    # Prediction
    y_pred = Models[model].predict(x_test)

    # Metrics
    CM = confusion_matrix(y_test, y_pred)
    F1 = f1_score(y_test, y_pred, average='micro')
    accuracy = accuracy_score(y_test, y_pred, normalize=False)
    loss = zero_one_loss(y_test, y_pred, normalize=False)
    recall = recall_score(y_test, y_pred, average='micro')
    precision = precision_score(y_test, y_pred, average='micro')

    # Classification Report
    report = classification_report(y_test, y_pred)

    # Score
    print('Training Score:', Models[model].score(x_train, y_train))
    print('Testing Score:', Models[model].score(x_test, y_test))

    # Update best model
    if best_score < F1:
        best_score = F1
        best_model = model

    # Print metrics
    print()
    print('Confusion Matrix\n', CM)
    print('Classification Report\n', report)  # Print classification report
    print('_________________________________________________')
    # print('F1 Score:', F1)
    # print('Number of Right Predictions:', accuracy)
    # print('Number of Wrong Predictions:', loss)
    # print('Recall Score:', recall)
    # print('Precision Score:', precision)
    # print('=================================================================')
```

```python
print('Best  Model:', best_model)
print('Best Score:', best_score)
```
